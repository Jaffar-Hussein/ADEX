# **Leaky Integrate and Fire Model with Spiking**

*This is the improvement on the LIF whereby we introduce an injection of a current to cause a spike to appear*

*Initialization*

```python
import matplotlib.pyplot as plt
# An abitrary chosen voltage
voltage = - 10
# The times it takes to decay tau
tau = 2.0
# the change in time we are dividing a second into a 100 steps
dt = 1.0/200
# The resting membrane potentiel
v_o = -70
# The time we are going to start the simulation 
t = 0
# The values of the voltage that changes are stored in a list for easy plotting
voltages = []
# Spike times are also counted
spike_times = []
# Threshold value - an abitary value for the voltage when achieved a spike is recorded
thresh = 20
```

*The code*

```python
while t < 70:
    voltages.append(voltage)

    dv = -(voltage - v_o)/tau
    # The spike 
    if t > 5 :
        dv += 50
        spike_times.append(t)
    voltage += dv * dt
        
    # The reset when it reaches our threshold
    if voltage >= thresh:
        voltage = v_o
       
    t += dt
# print(voltages)
```

*Plotting*

```python
plt.plot(voltages)
plt.show()
```

```python
plt.plot(spike_times)
plt.show()
```
